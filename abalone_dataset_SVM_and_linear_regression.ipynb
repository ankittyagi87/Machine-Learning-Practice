```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import mean_squared_error

col = ["sex", "length", "diameter", "height", "whole_weight", "shucked_weight", "visera_weight", "shell_weight", "rings"]
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data",names=col)
df['sex']
dummies = pd.get_dummies(df['sex'])
df = pd.concat([df, dummies], axis=1)
print(df.head())
print(df.describe())
print(df.info())
#plt.plot(df['rings'], df['length'], linestyle='none', marker='.')
#plt.show()
#sns.pairplot(df)
#plt.show()

print(df.corr())
X = df.drop(['rings', 'sex'], axis=1)
y = df['rings']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=121)
print(X_train.shape, X_test.shape)
lm = LinearRegression()
lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
print(lm.score(X_test, y_test))
print(mean_squared_error(y_test, predictions))

#alphas = np.linspace(0.1, 1.0, 10)

#for alpha in alphas:
#   ls = Lasso(alpha)
#   ls.fit(X_train, y_train)
#   predictions = ls.predict(X_test)
#   print(alpha, ls.score(X_test, y_test))
#print(mean_squared_error(y_test, predictions))

svr = SVR()
parameters = {'kernel': ['linear', 'rbf', 'poly'], 'C':[1000]}
clf = GridSearchCV(svr, parameters, cv=5)
clf.fit(X_train,y_train)
#print(clf.degree_)
print(clf.best_params_)
print(clf.score(X_test, y_test))

for k in range(30):
    k=k+1
    knn = KNeighborsRegressor(k)
    knn.fit(X_train, y_train)
    print(k, knn.score(X_test, y_test))
```
